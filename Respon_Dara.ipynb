```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
```

```python
image = plt.imread('respon.jpg')
plt.imshow(image)
```

```python
tinggi, lebar = image.shape[:2]
redImage = np.zeros((tinggi,lebar,3)).astype(int)
greenImage = np.zeros((tinggi,lebar,3)).astype(int)
blueImage = np.zeros((tinggi,lebar,3)).astype(int)

for i in range(tinggi):
	for j in range(lebar):
		(r, g, b) = image[i,j]
		redImage[i,j] = [255, r, r]
		greenImage[i,j] = [g, 255, g]
		blueImage[i,j] = [b, b, 255]

plt.figure( figsize = (15,15))
plt.subplot(131)
plt.title('Red Image')
plt.imshow(redImage)
plt.subplot(132)
plt.title('Green Image')
plt.imshow(greenImage)
plt.subplot(133)
plt.title('Blue Image')
plt.imshow(blueImage)
plt.show()
```

```python
def image_zooming(image, scale_factor):
    height, width = image.shape[:2]

    new_height = int(height * scale_factor)
    new_width = int(width * scale_factor)
    scaled_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

    return scaled_image

grayImage = cv2.imread('respon.jpg', cv2.IMREAD_GRAYSCALE)
scale_factor = 2 #coba 2
zoomed_image = image_zooming(grayImage,scale_factor)

plt.figure(figsize=(15,15))

plt.subplot(1,2,1)
plt.imshow(grayImage, cmap='gray')
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(grayImage, cmap='gray')
plt.title('Zoomed Image')
plt.show()
```

```python
hist_original, bins_original = np.histogram(grayImage.flatten(), bins=256, range=[0,256])
hist_zoomed, bins_zoomed = np.histogram(zoomed_image.flatten(), bins=256, range=[0,256])


plt.figure(figsize=(15,15))

plt.subplot(1,2,1)
plt.bar(np.arange(256), hist_original, color='black')
plt.xlabel('Intensity')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.bar(np.arange(256), hist_original, color='black')
plt.title('Zoomed Image')
plt.xlabel('Intensity')
plt.ylabel('Frequency')
```
